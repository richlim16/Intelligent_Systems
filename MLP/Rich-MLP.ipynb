```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('diabetes_data.csv')
dataset
```

```python
X = dataset.iloc[:, 0:].values # specifying -1 leaves out one less column, not specifying means start from 0
y = dataset.iloc[:, -1].values

print(X[0])
print(dataset.columns)
```

```python
pd.DataFrame(X, columns = ['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history','bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes'])
```

```python
dCount = 0
ndCount = 0

for el in X:
    if(el[8] == 0):
        ndCount += 1
    else:
        dCount += 1

print("No diabetes: ", ndCount)
print("Diabetes : ", dCount)
```

```python
pd.DataFrame(y, columns= ["Diabetes"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
print(X[0])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) #error here
X_test = sc.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
length = len(X_train[0])

# X_train = le.fit_transform(X_train)
# X_test = le.transform(X_test)

for i in range(length):
    X_train[: ,i] = le.fit_transform(X_train[: ,i])

length2 = len(X_test[0])
for i in range(length2):
    X_test[:,i] = le.fit_transform(X_test[:,i])
```

```python
ann = tf.keras.models.Sequential()

#adding the input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#insert another hidden layer
# ann.add(tf.keras.layers.Dense(units=6, activation='softmax'))
#output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# ann.compile(optimizer = 'Lion', loss = 'binary_crossentropy', metrics = ['accuracy'])
```
