```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('diabetes_data.csv')
dataset
```

```python
X = dataset.iloc[:, 0:].values # specifying -1 leaves out one less column, not specifying means start from 0
y = dataset.iloc[:, -1].values

print(X[0])
print(dataset.columns)
```

```python
pd.DataFrame(X, columns = ['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history','bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes'])
```

```python
pd.DataFrame(y, columns= ["Diabetes"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
print(X[0])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_test[0]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) #error here
X_test = sc.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
length = len(X_train[0])

# X_train = le.fit_transform(X_train)
# X_test = le.transform(X_test)

for i in range(length):
    X_train[: ,i] = le.fit_transform(X_train[: ,i])

length2 = len(X_test[0])
for i in range(length2):
    X_test[:,i] = le.fit_transform(X_test[:,i])
```

```python
ann = tf.keras.models.Sequential()

#adding the input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
X_train = np.asarray(X_train).astype('float32')
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```
